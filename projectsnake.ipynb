```python
# CREATE THE SNAKE
import pygame
import time
import random
pygame.init() # initilises pygame modules for you 

display_width = 400
display_height  = 400

display = pygame.display.set_mode((display_width,display_height)) # creats a display surface where the fist argument is size (x,y)
# 400, 400 is about average i think 
pygame.display.update() # will make the suface appear on the users screen update with no input will update the whole screen
pygame.display.set_caption('SNAKE!') # sets the caption on the surface (window) that opens 

# for the RGB - red green blue - colour scheme
blue=(0,0,255)
red=(255,0,0)
white = (255,255,255) # gives the white colour
black = (0,0,0) # gives the black colour
green = (0,255,0) # gives the colour green
colour = (100,200,100) # somehow gives green 
magenta = ((255,0,230))

# game_over=False



 
snake_block=10
 

clock = pygame.time.Clock()
snake_speed= 15
font_style = pygame.font.SysFont(None, 25)

def Your_score(score):
    value = font_style.render("Your Score: " + str(score), True, black)
    display.blit(value, [0, 0])

def message(message,colour):
    message1 = font_style.render(message, True, colour) # draws text on a new page
    display.blit(message1, [(display_width)/4 , (display_height)/2]) # overlaps the message page with the display

def our_snake(snake_block, snake_list):
    for x in snake_list:
        pygame.draw.rect(display, black, [x[0], x[1], snake_block, snake_block])

def gameLoop():  
    game_over = False
    game_close = False
 
    x1 = display_width/2 # where the snake starts 
    y1 = display_height/2

    x1_change = 0
    y1_change = 0

    snake_List = []
    Length_of_snake = 1
    
    foodx = round(random.randrange(0, display_width - snake_block) / 10.0) * 10.0
    foody = round(random.randrange(0, display_width - snake_block) / 10.0) * 10.0
 
    while not game_over:
        
        while game_close == True:
                display.fill(white)
                message("GAME OVER \n Press Q-Quit or C-Play Again", red)
                pygame.display.update()

                for event in pygame.event.get():
                    if event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_q:
                            game_over = True
                            game_close = False
                        if event.key == pygame.K_c:
                            gameLoop()
# while not game_over:
#     for event in pygame.event.get(): # gets all events and removes them from the queue 
#         if event.type==pygame.QUIT: # QUIT is an event that allows you to quit the game so when the event made is quit it 
#             # will quit the game 
#             game_over=True 
#     pygame.draw.rect(display,blue,[250,150,1000,10]) # draws a rectangle .rect(surface to put it on, colour of rect, rect to 
    # draw - position and dimensions) - for colour you have to define it as above with the blue = (0,0,255)
#     pygame.draw.rect(screen, [red, blue, green], [left, top, width, height], filled) - coordinates need to be within the 
# surface size to be seen on the screen (surface) - [200,150,10,10] is just a square in the centre 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN: # pygame keydown allows use of the keyboard arrows to control the game
                if event.key == pygame.K_LEFT:
                    x1_change = -snake_block
                    y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = snake_block
                    y1_change = 0
                elif event.key == pygame.K_UP:
                    y1_change = -snake_block
                    x1_change = 0
                elif event.key == pygame.K_DOWN:
                    y1_change = snake_block
                    x1_change = 0
    
        if x1 >= display_width or x1 < 0 or y1 >= display_height or y1 < 0:
            message("Game Over", black)
            game_over = True # if the snake hits the perimeter of the display game over
 
        x1 += x1_change # if an arrow is pressed this is an event which will move the snake one direction or another
        y1 += y1_change
        display.fill(white)
        pygame.draw.rect(display, blue, [foodx, foody, snake_block, snake_block])
#         pygame.draw.rect(display, green, [x1, y1, snake_block, snake_block])
        snake_Head = []
        snake_Head.append(x1)
        snake_Head.append(y1)
        snake_List.append(snake_Head)
        if len(snake_List) > Length_of_snake:
            del snake_List[0]
 
        for x in snake_List[:-1]:
            if x == snake_Head:
                game_close = True
 
        our_snake(snake_block, snake_List)
        Your_score(Length_of_snake - 1)
 
 
        pygame.display.update()
 
        if x1 == foodx and y1 == foody:
            foodx = round(random.randrange(0, display_width - snake_block) / 10.0) * 10.0
            foody = round(random.randrange(0, display_height - snake_block) / 10.0) * 10.0
            Length_of_snake += 1
#         pygame.display.update()
 
        clock.tick(snake_speed) # sets the speed of the game so 1 is very slow and 100 is super fast   

# message("Game Over", white)
# pygame.display.update()
# time.sleep(2) # not sure what this does

    
    pygame.quit() # signs out of pygame 
    quit()
    
gameLoop()
```
